```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
import time
import pandas as pd

PATH = r"C:\Users\vedpr\Downloads\chromedriver.exe"
driver = webdriver.Chrome(PATH)

driver.get("https://open.spotify.com/")
driver.find_element(By.XPATH,'//*[@id="main"]/div/div[2]/div[1]/header/div[5]/button[2]').click()

id=""
password=""

driver.implicitly_wait(15)

try:
    email = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="login-username"]')))
except:
    print("Email login not found")
    driver.quit()

email.clear()
email.send_keys('')
password = driver.find_element(By.XPATH,'//*[@id="login-password"]')

password.send_keys("")

driver.find_element(By.XPATH,'//*[@id="login-button"]/div[1]').click()

WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[4]/div/div[2]/nav/div[1]/div[2]/div/div[2]/a/span'))).click()
    
time.sleep(10)

ls = driver.find_elements(By.XPATH,'//div[@class="ShMHCGsT93epRGdxJp2w Ss6hr6HYpN4wjHJ9GHmi"]')

e = ['E'] #Spotify gives 'e' tags to explicit songs
for items in ls:
    songComponents = items.text.split('\n')
    
for n in e:
    while n in songComponents:
        songComponents.remove(n)

songComponents = songComponents[4:]
chunks = [songComponents[x:x+6] for x in range(0, len(songComponents), 6)]

driver.close()

df = pd.DataFrame(chunks,columns=['S.N.','Song','Artists','Album','Date Added','Duration'])
df = df.set_index('S.N.')
song = df['Song'].tolist()
```
